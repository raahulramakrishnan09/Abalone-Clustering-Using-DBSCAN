```python
#Label_encoding
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
df['Sex']=a.fit_transform(df['Sex'])
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=1,min_samples=240,metric='euclidean')
```

```python
model=db.fit(df)
```

```python
label=model.labels_
label
```

```python
#n_clusters
len(set(label))-(1 if -1 in label else 0)
```

```python
y_means=db.fit_predict(df)
y_means
```

```python
x=df.iloc[:,[5,6]].values
x
```

```python
#Visualizing_the_Clusters
plt.figure(figsize=(7,5))
plt.scatter(x[y_means==0,0],x[y_means==0,1],color='red',label='cluster1',s=30)
plt.scatter(x[y_means==1,0],x[y_means==1,1],color='yellow',label='cluster2',s=30)
plt.scatter(x[y_means==2,0],x[y_means==2,1],color='green',label='cluster3',s=30)
plt.scatter(x[y_means==3,0],x[y_means==3,1],color='blue',label='cluster5',s=30)
plt.scatter(x[y_means==4,0],x[y_means==4,1],color='orange',label='cluster6',s=30)
plt.scatter(x[y_means==5,0],x[y_means==5,1],color='pink',label='cluster7',s=30)
plt.scatter(x[y_means==6,0],x[y_means==6,1],color='indigo',label='cluster8',s=30)
plt.grid(False)
plt.title('Clustering')
plt.xlabel('Diameter')
plt.ylabel('rings')
plt.show()
```
